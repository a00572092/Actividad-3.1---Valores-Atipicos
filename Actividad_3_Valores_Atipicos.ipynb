```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_excel("/content/Gastos y costos 20-23.xlsx", sheet_name="2023", header=3)
df
```

```python
df.info()
```

```python
nulos = df.isnull().sum()
nulos
```

```python
#Sustitución de valores nulos por valores no nulos hacia atrás backward fill" ("bfill")
fol = df["Folio"] =df["Folio"].fillna(method="ffill")
mp = df["MP"] =df["MP"].fillna(method="ffill")
fp = df["FP"] =df["FP"].fillna(method="ffill")
tg = df["TIPO GASTO"] =df["TIPO GASTO"].fillna(method="bfill")
p = df["Poliza"] =df["Poliza"].fillna(method="bfill")
df = df.drop("Otros", axis=1)
```

```python
#Realizamos diagrama de caja
fig = plt.figure(figsize =(20, 20))
df.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show()
```

```python
#Método desviación estandar para eliminar los outliers6
y = df
limite_superior = y.mean() + 3*y.std()
limite_inferior = y.mean() - 3*y.std()
print("Límite superior permitido" , limite_superior)
print("\nLímite inferior permitido" , limite_inferior)
```

```python
#Conversion atipico-nulo
df2 = df[(y <= limite_superior) & (y >=limite_inferior)]
df2
df3 = df2.drop(["Fecha", "Folio", "UUID", "RFC","Proveedor", "TIPO GASTO", "Descripción","MP", "Tipo", "Status", "Poliza"], axis=1)
df3
```

```python
#Corroboración los valores nulos
df3.isnull().sum()
```

```python
#Sustitución con valores de media
dflimp = df3.copy()
dflimp = dflimp.fillna(round(df3.mean(),1))
dflimp.isnull().sum()
```

```python
plt.boxplot(dflimp)
plt.suptitle("Outliers con procedimiento de desviación estandar")
plt.show()
```

```python
#Método de rango intercuantil
y = dflimp
per25 = y.quantile(0.25) #Q1
per75 = y.quantile(0.75) #Q3
iqr = per75 - per75
lim_sup_iqr = per75+ 1.5*iqr
lim_inf_iqr = per25- 1.5*iqr
print("Límite superior permitido" , lim_sup_iqr)
print("\nLímite inferior permitido" , lim_inf_iqr)
```

```python
#Datos limpios del df
df3_iqr = dflimp[(y <=lim_sup_iqr)&(y >=lim_inf_iqr)]
df3_iqr
```

```python
nulos = df3_iqr.isnull().sum()
nulos
```

```python
#Reemplazo valores atípicos
df5_iqr = df3_iqr.copy()
df5_iqr = df5_iqr.fillna(round(df3_iqr.mean(),1))
df5_iqr
```

```python
#Se concatenan las filas string removidas
columnas = [ "Folio", "UUID", "RFC","Proveedor", "TIPO GASTO", "Descripción","MP", "Tipo", "Status", "Poliza"]
indice= df[columnas]
datoslimp = pd.concat([indice, df5_iqr],axis=1)
datoslimp
```

```python
plt.boxplot(df5_iqr)
plt.suptitle("Outliers con procedimiento de rango intercuantil")
plt.show()
```
